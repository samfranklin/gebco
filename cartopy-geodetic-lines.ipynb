```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()
plt.show()
```

```python
fig1 = plt.figure(1,(20, 10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

lon_lon, lon_lat = -0.12, 51.5
sing_lon, sing_lat = 103.8, 1.4
tok_lon, tok_lat = 139.7, 35.7

plt.plot([lon_lon, sing_lon], [lon_lat, sing_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([lon_lon, sing_lon], [lon_lat, sing_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.plot([lon_lon, tok_lon], [lon_lat, tok_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([lon_lon, tok_lon], [lon_lat, tok_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.show()
```

```python
# set figure size
fig = plt.figure(1,(20, 15))

# set location coordinates
lon_lon, lon_lat = -0.12, 51.5
sing_lon, sing_lat = 103.8, 1.4
tok_lon, tok_lat = 139.7, 35.7

# plot figure 1
ax1 = fig.add_axes([0.1, 0.6, 0.4, 0.4], projection=ccrs.Orthographic(60, 50))
ax1.stock_img()
plt.plot([lon_lon, sing_lon], [lon_lat, sing_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )
plt.title('London to Singapore')

# plot figure 2
ax2 = fig.add_axes([0.45, 0.6, 0.4, 0.4], projection=ccrs.Orthographic(60, 70))
ax2.stock_img()
plt.plot([lon_lon, tok_lon], [lon_lat, tok_lat],
         color='red', linewidth=2, marker='o',
         transform=ccrs.Geodetic())
plt.title('London to Tokyo')

# plt.subplots_adjust(left=0.2, wspace=0.8, top=0.8)

plt.show();
```
