```python
import matplotlib.pyplot as plt
# %matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.figure()       # initialize a figure
plt.plot(x,y_sin)  # plot series of coordinates as a line
plt.plot(x,y_cos)
plt.xlabel('x')    # adds label to x axis
plt.ylabel('y')    # adds label to y axis
plt.title('title') # adds a title
plt.show()         # close a figure
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.despine()
```

```python
with sns.axes_style('darkgrid'):
    plt.figure()
    plt.plot(x,y_cos)
    plt.plot(x,y_sin)
    plt.show()
    sns.set()
```
